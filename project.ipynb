```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class retailSalesAnalsis:
   def __init__(self):
    self.data=pd.read_csv('retail_sales.csv')
    self.data['Date']=pd.to_datetime(self.data['Date'])

   def data_clean(self):
    self.data.dropna(inplace=True)

   def total_product_sale(self):
    return self.data.groupby('Product')['Sales'].sum()

   def best_sell_product(self):
    return self.total_product_sale().sort_values(ascending=False).index[0]

   def avg_daily_sale(self):
    return self.data['Sales'].mean()

   def plot_sale_trend(self):
    self.data.groupby('Date')['Sales'].sum().plot(kind='line')
    plt.xlabel('Date')
    plt.ylabel('Total Sale')
    plt.title('Retail Sales Trend')
    plt.show()

   def plot_sale_by_product(self):
    return self.total_product_sale().plot(kind='bar')
    plt.xlabel('Product')
    plt.ylabel('Total Sale')
    plt.title('Retail Sales per Product')
    plt.show()



```

```python
analyzer=retailSalesAnalsis()
print('Total sale per product:',analyzer.total_product_sale())
print('Best selling product:',analyzer.best_sell_product())
print('Average daily sale:',analyzer.avg_daily_sale())
analyzer.plot_sale_trend()
analyzer.plot_sale_by_product()
```
